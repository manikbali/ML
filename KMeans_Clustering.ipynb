```python
# -*- coding: utf-8 -*-
"""
Created on Sun Feb  9 21:48:00 2020

@author: Manik Bali
"""
#Problem 1.
#Implement the K-means algorithm over MATLAB, 
#and cluster the samples in bothdatasets usingC= 2,3,,8 clusters. 
#Plot the members of each cluster using dif-ferent symbols, e.g., 
#x, o, etc., forC= 2,4 and 8 only.
 
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

#input number of classes that you wish to divide your data into
#C= 2,3,,8 
classes=8


#Read in data1 data
data1 = np.array(sio.loadmat(r"C:\Users\Navya\Downloads\data1.mat")['X'])

#Flatten the data to simplify processing. Our goal is the find distance between data set and centroid
#So flattening simplifies solution algorithm 
a= data1.flatten()

c=np.array(range(11),dtype='f')
#Assign Random centroid values to start with

c[0]=data1[0,175] 
c[1]=data1[0,275] 
c[2]=data1[1,375] 
c[3]=data1[0,475] 
c[4]=data1[0,575] 
c[5]=data1[0,675] 
c[6]=data1[1,575] 
c[7]=data1[1,675] 
c[8]=data1[1,695]
c[9]=data1[0,695]
c[10]=data1[0,685]
#While counter initilization
j=0

while j < 103:    #Arbitary 103 but can be programmed to stop at convergence. 
#stack flattend data rows and construct an array ( a_stacked ) that has  rows equal to classes( of flattend data1)
#All rows are identical
#Number of rows = number of classes 
#This helps in calculating distance of each centroid from all the data in a single array operation ( Line 42)
    a_stacked=np.dstack([a.squeeze()]*classes).squeeze()
    dist=abs(a_stacked - c[0:classes]) # Distance Function simplified. Square root and under root cancel
#Loop over classes. For each class gather the points of min distance ( inner part of line 44). 
#Mean of collected data gives the new centroid
#Exctract Data is plotted once the series has converged
#np.argmin(dist,axis=1) == i) gathers indicies of a column that has min distance to a given centroid
    
    for i in range(0,classes):
         c[i]=np.mean(a[(np.where(np.argmin(dist,axis=1) == i))])  
         print(i,c[i],c[i])
         if(j==102):
             plt.plot(a[(np.where(np.argmin(dist,axis=1) == i))],label='Centroid='+str(c[i]))
    j=j+1
plt.title('K-Means Decomposition of data1 into ' +str(classes)+ ' classes') 
plt.legend(loc="upper Right")    
```
